```python
import urllib.request
import requests
from bs4 import BeautifulSoup
```

```python
def getdata(url):
    r = requests.get(url)
    return r.text
```

```python
page_num = 0
count = 0

for i in range(400):
    htmldata = getdata("https://myanimelist.net/character.php?limit="+str(page_num))
    soup = BeautifulSoup(htmldata, 'html.parser')
    divTag = soup.find_all("table", {"class": "characters-favorites-ranking-table"})

    for tag in divTag:
        trTags = tag.find_all("tr", {"class": "ranking-list"})

        for tag in trTags:
            tdTags = tag.find_all("td", {"class": "people"})

            for characters in tdTags:

                for images in characters.find_all("img"):
                    count = count + 1
                    imgURL = images['data-srcset'].split(", ")[1].split(" ")[0]

                    filename = 'character'+str(count)

                    urllib.request.urlretrieve(imgURL, "./Characters/"+str(filename))
    
    page_num = page_num + 50

        


```
